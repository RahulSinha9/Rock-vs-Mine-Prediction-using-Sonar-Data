```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('Copy of sonar data1.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
x = df.drop(columns = '61' , axis = 1)
x
```

```python
y = df['61']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30,stratify=y, random_state = 52)
```

```python
x.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
#Accuracy on training data
from sklearn.metrics import accuracy_score
```

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = (accuracy_score(x_train_prediction,y_train)*100)
```

```python
training_data_accuracy
```

```python
#Accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = (accuracy_score(x_test_prediction, y_test)*100)
```

```python
test_data_accuracy
```
